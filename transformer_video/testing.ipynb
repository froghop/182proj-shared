```python
import dataset

import numpy as np
import matplotlib.pyplot as plt
```

```python
ds = dataset.Dataset(shape_side_length=3, invert_colors=True)

samples=3
sequence_length=9

q = ds.query(samples=samples, sequence_length=sequence_length, as_tensor=True, seed=321)
# print(q)
seqs = q['sequences']
seqs[0].shape
```

```python
plt.figure(figsize=(2*sequence_length, 2*samples), constrained_layout=True)
for i in range(samples):
    for j in range(sequence_length):
        plt.subplot(samples, sequence_length, i*sequence_length+j+1)
        plt.imshow(seqs[i][j], cmap='gray')
```
